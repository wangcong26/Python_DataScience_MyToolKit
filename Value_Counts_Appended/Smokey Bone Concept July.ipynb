## Test merge value count tables

```python
q4a=concept.columns[1:57]
q4b=concept.columns[57:113]
q4c=concept.columns[113:169]
q4c2=concept.columns[169:225]
q4d=concept.columns[225:281]
q4e=concept.columns[281:337]
q5=concept.columns[337:]
listOfList=[q4a,q4b,q4c,q4c2,q4d,q4e]
```

```python
def freqdist(mylist, mydataframe):
    tFrameList=[]
    for each in mylist:
        table=mydataframe[each].value_counts(sort=False,normalize=True)
        tableFrame=table.to_frame()
        tFrameList.append(tableFrame)
    result = pd.concat(tFrameList, axis=1)
    return result
```

```python
# use the function to create output
myq4a=freqdist(q4a, concept)
myq4b=freqdist(q4b, concept)
myq4c=freqdist(q4c, concept)
myq4c2=freqdist(q4c2, concept)
myq4d=freqdist(q4d, concept)
myq4e=freqdist(q4e, concept)
myq5=freqdist(q5,concept)
```

```python
# write to excel
with pd.ExcelWriter('output.xlsx') as writer:
    myq4a.to_excel(writer, sheet_name='q4a')
    myq4b.to_excel(writer, sheet_name='q4b')
    myq4c.to_excel(writer, sheet_name='q4c')
    myq4c2.to_excel(writer, sheet_name='q4c2')
    myq4d.to_excel(writer, sheet_name='q4d')
    myq4e.to_excel(writer, sheet_name='q4e')
    myq5.to_excel(writer, sheet_name='q5')
```
